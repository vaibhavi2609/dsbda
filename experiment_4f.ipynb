```python

df_x=df[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'black', 'lstat']]
print(df_x)
```

```python

df_y=df.price
print(df_y)
```

```python

reg =linear_model.LinearRegression()
```

```python

x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.25, random_state=42)
```

```python

reg.fit(x_train, y_train)
```

```python

print(reg.coef_)
```

```python

y_pred = reg.predict(x_test)
print(y_pred)
```
